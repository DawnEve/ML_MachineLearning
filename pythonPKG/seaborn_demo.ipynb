## 整体风格设置

```python
def sinplot(flip=1):
    x=np.linspace(0,14, 100)
    for i in range(1,7):
        y=np.sin(x+i*0.5) *(7-i)*flip
        plt.plot(x, y)
sinplot()
```

## 产生大量区分度大的颜色

```python
def createColors(n):
    baseColors=['red','orange','brown', '#A8FF00','lime',
                '#12e193','green',
                'blue','#0d75f8','cyan','#FF00FF','purple']
    L=len(baseColors)
    numPerColor=(n // (2*L-1))+1;
    if numPerColor<2:
        numPerColor=2
    
    #重色、浅色
    def g(baseColor, i):
        c1=sns.dark_palette(baseColor, n_colors=i)
        c2=sns.light_palette(baseColor, n_colors=i, reverse=True)
        return c1[1:]+c2[1:-1]
    rs=[]
    arr2=[]
    i=0
    for c in baseColors:
        print(c,i)
        i+=1
        tmp=g(c, numPerColor)
        sns.palplot(tmp)
        plt.show()
        #
        rs = rs+tmp
        arr2.append(tmp)
        
    print(L,numPerColor, len(rs))
    return  (rs, arr2);
#

c3, arr=createColors(80)
sns.palplot(c3)
#print(c3)
```

```python
#手工筛选
my=arr[0]+arr[1]+ \
    [arr[2][2]] + \
    arr[3][0:3]+arr[3][4:] + \
    arr[4][1:3]+arr[4][4:] + \
    arr[5][1:3]+arr[5][4:] +\
    arr[6][1:]+\
    arr[7][1:]+\
    arr[8][2:4]+\
    arr[9][1:3]+arr[9][4:]+\
    arr[10][1:]+\
    arr[11][2:]


sns.palplot(my)
len(my)
```
