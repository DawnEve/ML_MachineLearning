# filter

```python
titanic.head()
```

```python
#一等舱的生存率
a=titanic.loc[titanic['pclass']==1, 'survived']; 
a.sum()/a.shape[0]
```

```python
#每个仓位的生存率
for i in range(1,4):
    a=titanic.loc[titanic['pclass']==i, 'survived']; 
    ratio=a.sum()/a.shape[0]
    print(i, ratio)
```

```python
#找出老人和孩子
a1=titanic.loc[ (titanic['age']<15) | (titanic['age']>60)]
print(titanic.shape[0], a1.shape) #(100, 15)
a1.head()
```

```python
# 找出在一等舱的孩子
a2=titanic.loc[ (titanic['age']<15) & (titanic['pclass']==1)]
print(a2.shape[0]) #5
a2.head()
```

```python
#生存率， 按照仓位、性别区分
titanic.pivot_table('survived', 'pclass','sex')
```

```python
a3=titanic.loc[titanic['who']=='child']
print(titanic.shape[0], a3.shape[0])
a3.head()
```

```python
titanic.pivot_table('survived', 'pclass','who')
```

```python
## 每个仓位中年龄最大的
for i in titanic['pclass'].unique():
    rs=titanic.loc[ (titanic['pclass']==i), 'age'].max()
    print(i,rs)
```
